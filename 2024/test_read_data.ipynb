```python
import pandas as pd
import numpy as np
```

```python
DATA_PATH = 'drive/MyDrive/march_madness/2024/'
```

```python
df_Mteams = pd.read_csv(DATA_PATH + 'march-machine-learning-mania-2024/MTeams.csv').drop(columns=['FirstD1Season', 'LastD1Season'])
df_Wteams = pd.read_csv(DATA_PATH + 'march-machine-learning-mania-2024/WTeams.csv')
df_Wteams
```

```python
df_SOS = pd.DataFrame()

for yr in range(2003, 2024):
  df = pd.read_html(f'https://www.sports-reference.com/cbb/seasons/women/{yr}-ratings.html#ratings')[0]
  df = df.droplevel(0, axis=1)[['School', 'SOS']].dropna()
  df = df[df['School'] != 'School'].reset_index(drop=True)
  df['Season'] = np.ones(df.shape[0], dtype=int)*yr
  df_SOS = pd.concat([df_SOS,df])

df_SOS = df_SOS.reset_index(drop=True)
print(df_SOS.shape)
df_SOS
```

```python
df_SOS = pd.read_csv(DATA_PATH + 'sports_ref_W2003.csv')[['School', 'SOS']]
df_SOS = df_SOS[df_SOS['School'] != 'School'].reset_index(drop=True)
df_SOS['Season'] = np.ones(df_SOS.shape[0], dtype=int)*2003
df_SOS
```

```python
df_WSOS = df_Wteams.merge(df_SOS, right_on='School', left_on='TeamName', how='outer')
df_WSOS[df_WSOS.SOS.isna()]
```

```python
df_SOS[df_SOS.School == 'Abilene Christian'].School = 'Abilene Chr'
```

```python
df_SOS.to_csv(DATA_PATH + 'WStrengthOfSchedule.csv')
```
